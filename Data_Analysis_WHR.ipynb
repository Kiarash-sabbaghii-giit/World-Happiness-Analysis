```python
def find_country_match(input_name, threshold=0.6):
    """
    Intelligent search to find the closest country name match
    """
    # Clean the input: convert to string, remove whitespace, lowercase
    input_name = str(input_name).strip().lower()
    
    # 1. Exact match search
    for country in pycountry.countries:
        if input_name == country.name.lower():
            return country.name, "exact"
    
    # 2. Partial match search (contains relationship)
    for country in pycountry.countries:
        country_lower = country.name.lower()
        # Check if either string contains the other
        if country_lower in input_name or input_name in country_lower:
            return country.name, "partial"
    
    # 3. Similarity search using difflib
    all_country_names = [c.name for c in pycountry.countries]
    matches = difflib.get_close_matches(
        input_name.title(),  # Convert to Title case for better accuracy
        all_country_names,
        n=1,  # Get only the best match
        cutoff=threshold  # Minimum similarity threshold (0-1)
    )
    
    if matches:
        return matches[0], "similar"
    
    # No match found
    return None, "no_match"

# Initialize results dictionary
results = {
    "exact_match": [],  # Store exact matches
    "partial_match": [],  # Store partial/contains matches
    "similar_match": [],  # Store similar matches
    "no_match": []  # Store no matches
}

# Process each country name in the data
for country_name in data['Country']:
    # Handle missing values
    if pd.isna(country_name):
        results["no_match"].append(("NaN/Null", None))
        continue
    
    # Find match using our function
    matched_name, match_type = find_country_match(country_name)
    
    # Categorize the result
    if matched_name:
        # Add to appropriate match category
        results[f"{match_type}_match"].append((country_name, matched_name))
    else:
        # No match found
        results["no_match"].append((country_name, None))

# Display results summary
print("=" * 60)
print("Country Name Analysis Results:")
print("=" * 60)

for match_type, items in results.items():
    print(f"\n{match_type.upper().replace('_', ' ')} ({len(items)} items):")
    
    if match_type == "no_match":
        # Display first 20 items with no match
        for orig, _ in items[:20]:
            print(f"   '{orig}'")
    else:
        # Display first 10 matched items
        for orig, matched in items[:10]:
            print(f"   '{orig}'  '{matched}'")
    
    # Show "and X more items" if there are more items
    display_limit = 20 if match_type == "no_match" else 10
    if len(items) > display_limit:
        remaining = len(items) - display_limit
        print(f"  ... and {remaining} more items")

# Print summary statistics
print("\n" + "=" * 60)
print("Summary:")
print(f"Exact matches: {len(results['exact_match'])} items")
print(f"Partial matches: {len(results['partial_match'])} items")
print(f"Similar matches: {len(results['similar_match'])} items")
print(f"No matches: {len(results['no_match'])} items")

# Optional: Create a new column in the dataframe with cleaned country names
def get_best_match(name):
    """Helper function to get the best matched country name"""
    matched_name, _ = find_country_match(name)
    return matched_name if matched_name else name

# Add cleaned country column to dataframe (optional)
data['Country_Cleaned'] = data['Country'].apply(get_best_match)

# Display sample of cleaned data
print("\n" + "=" * 60)
print("Sample of Cleaned Country Names:")
print("=" * 60)
sample_data = data[['Country', 'Country_Cleaned']].head(10)
print(sample_data.to_string())

# Additional analysis: Show most common issues
print("\n" + "=" * 60)
print("Most Common No-Match Entries:")
print("=" * 60)

# Count frequency of no-match entries
from collections import Counter
no_match_counts = Counter([item[0] for item in results['no_match'] if item[0] != "NaN/Null"])

if no_match_counts:
    for item, count in no_match_counts.most_common(10):
        print(f"  '{item}' appears {count} times")
else:
    print("  No issues found!")
```

```python
happiness_stats = (
    Data
    .groupby('Year')['Happiness_Score']
    .agg(
        Mean='mean',
        Median='median',
        Std='std'
    )
    .reset_index()
)

happiness_stats
```
